```python
def crea_grafo(tx):

    for i,(titolo,autore,genere) in enumerate(libri_reali):

        tx.run("""
        MERGE (a:Autore {nome:$autore})
        MERGE (s:Scaffale {genere:$genere})
        CREATE (l:Libro {id:$id, titolo:$titolo})

        MERGE (a)-[:HA_SCRITTO]->(l)
        MERGE (l)-[:NELLO_SCAFFALE]->(s)
        """,
        autore=autore,
        genere=genere,
        titolo=titolo,
        id=i+1
        )
```

```python
with driver.session() as session:
    session.execute_write(crea_grafo)

print("Database a grafo creato")
```

## SELECT (lettura dati)

```python
def mostra_libri(tx):
    result = tx.run("""
    MATCH (a:Autore)-[:HA_SCRITTO]->(l:Libro)
    RETURN a.nome, l.titolo
    """)
    for r in result:
        print(r)

with driver.session() as session:
    session.execute_read(mostra_libri)
```

## UPDATE

```python
def aggiorna_libro(tx):
    tx.run("""
    MATCH (l:Libro {id:1})
    SET l.titolo='Il Signore degli Anelli - Extended'
    """)

with driver.session() as session:
    session.execute_write(aggiorna_libro)

print("Libro aggiornato")
```

## DELETE

```python
def elimina_libro(tx):
    tx.run("""
    MATCH (l:Libro {id:1})
    DETACH DELETE l
    """)

with driver.session() as session:
    session.execute_write(elimina_libro)

print("Libro eliminato")
```

## Ricerca libro

```python
def cerca_libro(tx, titolo):
    result = tx.run("""
    MATCH (l:Libro)
    WHERE l.titolo CONTAINS $titolo
    RETURN l
    """, titolo=titolo)

    return list(result)

with driver.session() as session:
    libri = session.execute_read(cerca_libro, "Anelli")

print(libri)
```

```python
generi = {
"Guida galattica per autostoppisti":"SciFi",
"Lo Hobbit":"Fantasy",
"Arthur e i Minimei":"Fantasy",
"Io robot":"SciFi",
"L'enigma dei buchi neri":"Scienza",
"Buchi neri e salti temporali":"Scienza",
"In un volo di storni":"Scienza"
}
```

```python
def crea_libri_preferiti(tx):

    for libro in libri_preferiti:

        tx.run("""
        MERGE (l:Libro {titolo:$titolo})
        SET l.preferito = true
        """, titolo=libro)


with driver.session() as session:
    session.execute_write(crea_libri_preferiti)
```

```python
def crea_generi(tx):

    for libro,genere in generi.items():

        tx.run("""
        MERGE (l:Libro {titolo:$libro})
        MERGE (g:Genere {nome:$genere})

        MERGE (l)-[:NEL_GENERE]->(g)
        """,
        libro=libro,
        genere=genere
        )

with driver.session() as session:
    session.execute_write(crea_generi)
```

```python
similarita = [
("Guida galattica per autostoppisti","Io robot"),
("Io robot","Guida galattica per autostoppisti"),
("Lo Hobbit","Arthur e i Minimei"),
("L'enigma dei buchi neri","Buchi neri e salti temporali")
]
```

```python
def crea_similarita(tx):

    for l1,l2 in similarita:

        tx.run("""
        MATCH (a:Libro {titolo:$l1})
        MATCH (b:Libro {titolo:$l2})

        MERGE (a)-[:SIMILE_A]->(b)
        """,
        l1=l1,
        l2=l2
        )

with driver.session() as session:
    session.execute_write(crea_similarita)
```
